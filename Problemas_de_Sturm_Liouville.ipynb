1.$X'' + \lambda X = 0 $   
 $X(0)=X(1)=0$

Importamos las librerias necesarias

```python
from sympy import *
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

Definimos

```python
x = sp.symbols('x')
k = sp.pi**2  # valor de k elegido
X = sp.Function('X')
C1, C2 = sp.symbols('C1 C2')
```

Ecuacion Diferencial Ordinaria

```python
ode = sp.Eq(X(x).diff(x, 2) + k*X(x), 0)
```

Resolviendo la EDO

```python
sol1 = sp.dsolve(ode, X(x))
Xsol1 = sol1.rhs
sol1
```

Aplicando las condiciones

```python
conds1 = [
    Xsol1.subs(x, 0),
    Xsol1.subs(x, 1)
]
sol_c1 = sp.solve(conds1, (C1, C2))
Xsol1 = Xsol1 = Xsol1.subs(sol_c1).subs(C1, 1) # C2 es la constante libre

Xsol1
```

Funcion para graficar

```python
func1 = sp.lambdify(x, Xsol1, 'numpy')
xx = np.linspace(0, 1, 200)
yy1 = func1(xx)
```

Grafica

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.plot(xx, yy1, 'r', label='Problema 1')
plt.title("X'' + kX = 0, X(0)=X(1)=0")
plt.xlabel('x')
plt.ylabel('X(x)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

2.$X'' + \lambda X = 0 $   
 $X'(0)=X'(1)=0$

resolviendo la EDO

```python
sol2 = sp.dsolve(ode, X(x))
Xsol2 = sol2.rhs
dXsol2 = Xsol2.diff(x)
sol2
```

Aplicando las condiciones inicilaes

```python
conds2 = [
    dXsol2.subs(x, 0),  # X'(0) = 0
    dXsol2.subs(x, 1)   # X'(1) = 0
]
sol_c2 = sp.solve(conds2, (C1, C2))
# Sustituir soluciones y asignar valor a la constante restante
Xsol2 = Xsol2.subs(sol_c2).subs(C2, 1)  # C1 es la constante libre
Xsol2
```

Funcion para graficar

```python
func2 = sp.lambdify(x, Xsol2, 'numpy')
xx = np.linspace(0, 2, 200)
yy2 = func2(xx)
```

Grafica

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,2)
plt.plot(xx, yy2, 'b', label='Problema 2')
plt.title("X'' + kX = 0, X'(0)=X'(1)=0")
plt.xlabel('x')
plt.ylabel('X(x)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
