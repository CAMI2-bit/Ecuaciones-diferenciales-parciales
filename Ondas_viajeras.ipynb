####Onda Viajeras

Grafica la funcion

\begin{equation}
f(x') = exp  (-\frac{(x'-0.5)^2}{0.01} )
\end{equation}

y suponga que esta representa una perturbacion en una cuerda

Importamos las librerias necesarias

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

Definimos las variables necesarias

```python
x, t, c = sp.symbols('x t c')
```

Definimos la funcion

$f(x') = exp  (-\frac{(x'-0.5)^2}{0.01} )$

```python
x_prime = sp.symbols("x'")
f = sp.exp(-((x_prime - 0.5)**2) / 0.01)
```

(a) Mostrar que al hacer el cambio de variables $x'=x-ct$ la perturbacion se desplaza a la derecha

```python
x1 = x - c*t
f_a = f.subs(x_prime, x1)
print("Onda viajera a la derecha (a):")
sp.pprint(f_a)
```

Grafica del inciso (a)

```python
f_a_func = sp.lambdify((x, t), f_a.subs(c, 1), 'numpy')
x_vals = np.linspace(-1, 2, 500)
t_val = 0.2
y_a = f_a_func(x_vals, t_val)

plt.figure(figsize=(8, 4))
plt.plot(x_vals, y_a, label='(a) Onda a la derecha', color='blue')
plt.title("Onda viajera a la derecha (a)")
plt.xlabel("x")
plt.ylabel("f(x,t)")
plt.legend()
plt.grid(True)
plt.show()
```

(b) Mostrar que al hacer el cambio de variables $x'=x+ct$ la perturbacion se desplaza a la izquierda

```python
x2 = x + c*t
f_b = f.subs(x_prime, x2)
print("\nOnda viajera a la izquierda (b):")
sp.pprint(f_b)
```

Grafica del inciso (b)

```python
f_b_func = sp.lambdify((x, t), f_b.subs(c, 1), 'numpy')
y_b = f_b_func(x_vals, t_val)

plt.figure(figsize=(8, 4))
plt.plot(x_vals, y_b, label='(b) Onda a la izquierda', color='green')
plt.title("Onda viajera a la izquierda (b)")
plt.xlabel("x")
plt.ylabel("f(x,t)")
plt.legend()
plt.grid(True)
plt.show()
```

(c) Graficar la superposicion de las ondas viajeras de los incisos a) y b)

```python
f_c = f_a + f_b
print("\nSuperposición de ondas viajeras (c):")
sp.pprint(f_c)
```

Grafica del inciso (c)

```python
f_c_func = sp.lambdify((x, t), f_c.subs(c, 1), 'numpy')
y_c = f_c_func(x_vals, t_val)

plt.figure(figsize=(8, 4))
plt.plot(x_vals, y_c, label='(c) Superposición', color='red', linestyle='--')
plt.title("Superposición de ondas viajeras (c)")
plt.xlabel("x")
plt.ylabel("f(x,t)")
plt.legend()
plt.grid(True)
plt.show()
```
