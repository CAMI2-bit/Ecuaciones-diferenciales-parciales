Ecuacion log ́ıstica: $dp/dt=ap-bp^2, p(0)=p_0$
El problema de valor inicial para la ecuaci ́on log ́ıstica es ampliamente usado como un modelo crecimiento poblacional.Dondeaybson los coeficientes vitales de la poblaci ́onp(t) y $p_0$ es la poblaci ́on inicial.

```python
import sympy as sp #Primeramente importamos la librer ́ıa SymPy a Python
```

```python
#Definimos las variables independiente y dependiente as ́ı como los par ́ametros de forma simb ́olica
t=sp.symbols('t')
p=sp.Function('p(t)')
a=sp.symbols('a')
b=sp.symbols('b')
p0=sp.symbols('p0')
```

```python
#Ahora definimos la ecuaci ́on diferencial y la resolvemos con la condici ́on inicial usando dsolve
ode=sp.Eq(sp.Derivative(p(t),t)-a*p(t)+b*p(t)*p(t),0)
sol = sp.dsolve(ode,p(t),ics={p(0):p0})
```

```python
#Obteniendo as ́ı la soluci ́on:
sp.Eq(p(t), a*sp.exp(a*(t + sp.log(b*p0/(-a + b*p0))/a))/(b*(sp.exp(a*(t + sp.log(b*p0/(-a + b*p0))/a)) - 1)))
```

```python
#Definimos a continuaci ́on los valores de a, b y p0, y evaluamos las respectivas soluciones (una para cada condici ́on inicial):
constants={a:0.03,b:0.0001,p0:5}
sol1=sol.subs(constants)
constants={a:0.03,b:0.0001,p0:6}
sol2=sol.subs(constants)
constants={a:0.03,b:0.0001,p0:7}
sol3=sol.subs(constants)
```

```python
#La funci ́onlambdifypermite definir una funci ́on a partir de la expresi ́on simb ́olica:
func1 = sp.lambdify(t,sol1.rhs,'numpy')
func2 = sp.lambdify(t,sol2.rhs,'numpy')
func3 = sp.lambdify(t,sol3.rhs,'numpy')
```

```python
#A continuaci ́on, importamos los m ́odulosscipy, matplotlibypyplotpara graficar nuestras soluciones:
import scipy as sc
import numpy as np
tt = np.arange(0,200,1)
pp1 = func1(tt)
pp2 = func2(tt)
pp3 = func3(tt)
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.plot(tt,pp1,'r',tt,pp2,'b',tt,pp3,'g')
plt.xlabel('tiempo t')
plt.ylabel('Poblacion p(t)')
plt.show()
```

3. Resonancia: $y''+16y=8sen(4t), \;\;  y(0)=1, \;\; y'(0)=0$

```python
from sympy import *
import sympy as sp
t=sp.symbols('t')
y=sp.Function('y')
ode3=sp.Eq(sp.Derivative(y(t),t,t)+16*y(t)-8*sin(4*t),0)
```

```python
#Resolvemos la ecuaci ́on diferencial:
sol3 = sp.dsolve(ode3,y(t))
```

```python
#para obtener:
Eq(y(t), C2*sin(4*t) + (C1 - t)*cos(4*t))
```

```python
#Vamos ahora a definir las condiciones iniciales y substituirlas en la solucion:
const=sp.solve([sol3.rhs.subs(t,0)-1, sol3.rhs.diff(t,1).subs(t,0)-0])
C1, C2 = sp.symbols('C1,C2')
sol3=sol3.subs(const)
```

```python
#Definimos una funcion a partir de la solucion simb ́olica obtenida:
func1 = sp.lambdify(t,sol3.rhs,'numpy')
```

```python
import scipy
import numpy as np
xx = np.arange(0,10,.1)
yy1 = func1(xx)
```

```python
# Y graficamos con los comandos:
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.plot(xx,yy1,'r')
plt.xlabel('tiempo t')
plt.ylabel('y(t)')
plt.show()
```

6. Un problema de valor inicial con deltas de Dirac como fuentes:$y''(t)-4y'(t) + 4y(t) = 3δ(t-1) +δ(t-2), \;\; y(0) = 1, \;\; y'(0) = 1$

```python
from sympy import *
import sympy as sp
t=sp.symbols('t')
y=sp.Function('y')
ode6=sp.Eq(sp.Derivative(y(t),t,2)-4*sp.Derivative(y(t),t)+4*y(t),3*DiracDelta(t-1)+DiracDelta(t-2))
sol6 = sp.dsolve(ode6)
constants=solve([sol6.rhs.subs(t,0)-1,sol6.rhs.diff(t,1).subs(t,0)-1])
sol6.subs(constants)
```

```python
#y obtenemos
Eq(y(t), (t*(exp(-4)*Heaviside(t-2)+3*exp(-2)*Heaviside(t-1)-1)-2*exp(-4)*Heaviside(t-2) - 3*exp(-2)*Heaviside(t-1)+1)*exp(2*t))
```

```python
# Convertir la solución a una función numérica
f_lambdified = lambdify(t, sol6.rhs, 'numpy')

# Generar valores para la gráfica
t_vals = np.linspace(0, 5, 400)
y_vals = f_lambdified(t_vals)

# Graficar
plt.figure(figsize=(8, 5))
plt.plot(t_vals, y_vals, label='$y(t)$', color='b')
plt.xlabel('Tiempo $t$')
plt.ylabel('Respuesta $y(t)$')
plt.title('Solución de la ecuación diferencial')
plt.legend()
plt.grid()
plt.show()
```
