```python

import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from sympy.abc import x

#Generar los polinomios de Legendre usando sympy
legendre_polynomials = [sp.legendre(n, x) for n in range(6)]


x_vals = np.linspace(-1, 1, 400)
polys_evaluated = [sp.lambdify(x, Pn, modules=["numpy"])(x_vals) for Pn in legendre_polynomials]


plt.figure(figsize=(10, 6))

plt.title("Polinomios de Legendre $P_n(x)$ para $n=0$ a $5$")
plt.xlabel("$x$")
plt.ylabel("$P_n(x)$")
plt.grid(True)
plt.legend()
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from sympy.abc import x

#Polinomios de Legendre para n=3 y n=4
Pn = sp.legendre(3, x)
Pn1 = sp.legendre(4, x)

#derivada de Pn
dPn = sp.diff(Pn, x)

#Lado izquierdo y derecho de la identidad
lhs = (1 - x**2) * dPn
rhs = 4 * x * Pn - 4 * Pn1

#Lambdificar para evaluar
lhs_func = sp.lambdify(x, lhs, modules=["numpy"])
rhs_func = sp.lambdify(x, rhs, modules=["numpy"])

#Valores de x
x_vals = np.linspace(-1, 1, 400)
lhs_vals = lhs_func(x_vals)
rhs_vals = rhs_func(x_vals)


plt.figure(figsize=(10, 6))
plt.plot(x_vals, lhs_vals, label="Lado Izquierdo", linestyle='--')
plt.plot(x_vals, rhs_vals, label="Lado Derecho", linestyle='-')
plt.title("Verificación Gráfica de la Relación para n=3")
plt.xlabel("x")
plt.ylabel("Valor")
plt.legend()
plt.grid(True)
plt.show()
```

```python

# Verificación algebraica
simplified = sp.simplify(lhs - rhs)
print("¿La expresión se simplifica a 0?", simplified == 0)
```

```python

import sympy as sp
from sympy.abc import x
from math import factorial

for n in range(6):
    # Fórmula de Rodrigues
    f = (x**2 - 1)**n
    dfn = sp.diff(f, x, n)
    rodrigues = sp.simplify(1 / (2**n * factorial(n)) * dfn)

    #Polinomio de Legendre con sympy
    legendre_poly = sp.legendre(n, x)

    #Verificamos si son iguales
    iguales = sp.simplify(rodrigues - legendre_poly) == 0

    print(f"n = {n}: ¿Coinciden? {iguales}")
```
