```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)  # Valores de x
y = x**2  # Función y = x^2

plt.plot(x, y)  # Graficar
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = x^2')
plt.grid()
plt.show()

```
